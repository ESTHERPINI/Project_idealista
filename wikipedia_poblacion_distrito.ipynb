  ## Población Madrileña por distritos

### Pandas read html

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
url="https://es.wikipedia.org/wiki/Demograf%C3%ADa_de_Madrid"
```

```python
table_MN = pd.read_html(url)
```

```python
print(f'Total tables: {len(table_MN)}')
```

```python
#visualizamos y la tabla que nos interesa es la primera
df = table_MN[0]
df.head()
```

```python
#eliminamos columna del total
df.drop([0], inplace=True)
```

```python
datatypes = df.dtypes 
  
datatypes
```

```python
df["Población"] = pd.to_numeric(df["Población"], errors='coerce')
```

```python
datatypes = df.dtypes 
  
datatypes
```

```python
plt.figure(figsize=(16,7))

#sns.countplot(x=df.district)
ax = sns.barplot(x="Distrito", y="Población", data=df, palette='pastel' )
ax.set_xticklabels(ax.get_xticklabels(),  rotation=40, ha="right",fontsize=11)
plt.tight_layout() 

# Chart and axes titles
ax.set_title('Población por distrito Madrid - wikipedia',fontsize = 18, fontweight='bold' )
ax.set_xlabel('Distrito', fontsize = 15)
ax.set_ylabel('Población en miles', fontsize = 15)


plt.show() 
```

```python
#Para poder mergear con la tabla de idealista comprobamos que el nombre de los distritos es igual en ambas tablas
district_idealista=pd.read_excel("../Project/output/tabla_district.xls")
district_idealista_list=list(district_idealista.district.unique())
ordenados_idealista=sorted(district_idealista_list)
len(ordenados_idealista)
```

```python
distrito_poblacion_list=list(df.Distrito)
ordenados_poblacion=sorted(distrito_poblacion_list)
len(ordenados_poblacion)
```

```python
comparacion = []
 
for item in ordenados_poblacion:
    if item not in ordenados_idealista:
        comparacion.append(item)
    
for item in comparacion: 
    print("Elemento no está o no es idéntico en ambas listas",'%s' % item)
```

```python
df['Distrito'] = df['Distrito'].replace('Fuencarral-El Pardo','Fuencarral')
df['Distrito'] = df['Distrito'].replace('Moncloa-Aravaca','Moncloa')
df['Distrito'] = df['Distrito'].replace('Salamanca','Barrio de Salamanca')
df['Distrito'] = df['Distrito'].replace('San Blas-Canillejas','San Blas')
distrito_poblacion_list=list(df.Distrito)
ordenados_poblacion=sorted(distrito_poblacion_list)
len(ordenados_poblacion)
```

```python
comparacion = []
 
for item in ordenados_poblacion:
    if item not in ordenados_idealista:
        comparacion.append(item)
    
for item in comparacion: 
    print("Elemento no está en ambas listas",'%s' % item)
```

```python
df.rename(columns = {'Distrito':'district', 'Población':'population'}, inplace = True) 
df.head()
```
