### Centros educativos list from munimadrid website

```python
url = "https://datos.madrid.es/egob/catalogo/212790-0-centros-educacion.json"
data=api_centros(url)
data
```

### json_normalize to dataframe

```python
colegios = data['@graph']
df = json_normalize(colegios)
df.head(3)
```

### drop not usefull columns
