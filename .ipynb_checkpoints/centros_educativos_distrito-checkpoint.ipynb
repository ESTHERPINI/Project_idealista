```python
import requests
import pandas as pd
import json
from pandas.io.json import json_normalize
```

```python
colegios = "https://datos.madrid.es/egob/catalogo/212790-0-centros-educacion.json"
r = requests.get(colegios)

data=r.json()
```

```python
with open('colegios.csv', 'w') as fp:
    json.dump(colegios, fp)
```
